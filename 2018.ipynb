```python
class EntityDataset:
#     def __init__(self, texts, ner1, ner2,pos):
    def __init__(self, texts, ner1, ner2):
        self.texts = texts
        self.ner1 = ner1
        self.ner2 = ner2
#         self.pos = pos
    def __len__(self):
        return len(self.texts)

    def __getitem__(self, item):
        text = self.texts[item]
        ner1 = self.ner1[item]
        ner2 = self.ner2[item]
#         pos = self.pos[item]

        ids = []
        target_ner1 = []
        target_ner2 =[]
#         target_pos=[]
        mask=[]
        token_type_ids=[]


        inputs = config.TOKENIZER.encode(
                text,
                add_special_tokens=False
            )
        input_len = len(inputs)
        padding_len =config.MAX_LEN - input_len
        ids.extend(inputs + [0]*padding_len )
        target_ner1.extend(ner1+ [8]*padding_len)
        target_ner2.extend(ner2 + [6]*padding_len)
#         target_pos.extend(pos + [2]*padding_len)


        mask.extend([1] * input_len + [0]*padding_len)
        token_type_ids.extend([0]*config.MAX_LEN)

        return {
            "ids": torch.tensor(ids, dtype=torch.long),
            "mask": torch.tensor(mask, dtype=torch.long),
            "token_type_ids": torch.tensor(token_type_ids, dtype=torch.long),
            "target_ner1": torch.tensor(target_ner1, dtype=torch.long),
            "target_ner2": torch.tensor(target_ner2, dtype=torch.long),
#             "target_pos": torch.tensor(target_pos, dtype=torch.long),
        }
```

```python
def process_data(df):
    enc_ner1 = preprocessing.LabelEncoder()
    enc_ner2 = preprocessing.LabelEncoder()
    enc_pos= preprocessing.LabelEncoder()

    df.loc[:, "NER1"] = enc_ner1.fit_transform(df["NER1"])
    df.loc[:, "NER2"] = enc_ner2.fit_transform(df["NER2"])
    df.loc[:, "POS"] = enc_pos.fit_transform(df["POS"])

    sentences = df.groupby("Sentences")["Word"].apply(list).values
    ner1 = df.groupby("Sentences")["NER1"].apply(list).values
    ner2 = df.groupby("Sentences")["NER2"].apply(list).values
    pos = df.groupby("Sentences")["POS"].apply(list).values
    return sentences, ner1, ner2,pos, enc_ner1, enc_ner2,enc_pos
```

```python
def train_fn(data_loader, model, optimizer,optimizer1, device, scheduler):
    model.train()
    ner1_losses = []
    ner2_losses = []

    predicted_ner1=[]
    real_ner1=[]
    predicted_ner2=[]
    real_ner2=[]



    for data in tqdm(data_loader, total=len(data_loader)):
        for k, v in data.items():
            data[k] = v.to(device)
        optimizer.zero_grad()
        optimizer1.zero_grad()
        prediction_ner1, prediction_ner2,loss_ner1,loss_ner2 = model(**data)
        # pro.append(p)




        loss_ner1.backward(retain_graph=True)
        loss_ner2.backward(retain_graph=True)

        for sublist_a, sublist_b in zip(prediction_ner1, data["target_ner1"]):
          n = len(sublist_a)
          first_n_values = sublist_b[:n]

          real_ner1.append(first_n_values)
          predicted_ner1.append(sublist_a)

        for sublist_a, sublist_b in zip(prediction_ner2, data["target_ner2"]):
          n = len(sublist_a)
          first_n_values = sublist_b[:n]

          real_ner2.append(first_n_values)
          predicted_ner2.append(sublist_a)


        optimizer.step()
        optimizer1.step()
        scheduler.step()

        ner1_losses.append(loss_ner1.item())
        ner2_losses.append(loss_ner2.item())

    avg_loss_ner1 = sum(ner1_losses) / len(ner1_losses)
    avg_loss_ner2 = sum(ner2_losses) / len(ner2_losses)

    y_true_ner1 = torch.tensor([label for sublist in real_ner1 for label in sublist]).cpu()
    del real_ner1
    y_pred_ner1 = torch.tensor([label for sublist in predicted_ner1 for label in sublist]).cpu()
    del predicted_ner1
    f1_ner1 = f1_score(y_true_ner1, y_pred_ner1, average='micro')

    y_true_ner2 = torch.tensor([label for sublist in real_ner2 for label in sublist]).cpu()
    del real_ner2
    y_pred_ner2 = torch.tensor([label for sublist in predicted_ner2 for label in sublist]).cpu()
    del predicted_ner2
    f1_ner2 = f1_score(y_true_ner2, y_pred_ner2, average='micro')


    return avg_loss_ner1, avg_loss_ner2,f1_ner1,f1_ner2

def eval_fn(data_loader, model, device):
    model.eval()
    ner1_losses = []
    ner2_losses = []

    predicted_ner1=[]
    real_ner1=[]
    predicted_ner2=[]
    real_ner2=[]



    with torch.no_grad():
      for data in tqdm(data_loader, total=len(data_loader)):
          for k, v in data.items():
              data[k] = v.to(device)

          prediction_ner1, prediction_ner2,loss_ner1,loss_ner2 = model(**data)

          ner1_losses.append(loss_ner1.item())
          ner2_losses.append(loss_ner2.item())
          for sublist_a, sublist_b in zip(prediction_ner1, data["target_ner1"]):
            n = len(sublist_a)
            first_n_values = sublist_b[:n]

            real_ner1.append(first_n_values)
            predicted_ner1.append(sublist_a)

          for sublist_a, sublist_b in zip(prediction_ner2, data["target_ner2"]):
            n = len(sublist_a)
            first_n_values = sublist_b[:n]

            real_ner2.append(first_n_values)
            predicted_ner2.append(sublist_a)

      avg_loss_ner1 = sum(ner1_losses) / len(ner1_losses)
      avg_loss_ner2 = sum(ner2_losses) / len(ner2_losses)

      y_true_ner1 = torch.tensor([label for sublist in real_ner1 for label in sublist]).cpu()
      del real_ner1
      y_pred_ner1 = torch.tensor([label for sublist in predicted_ner1 for label in sublist]).cpu()
      del predicted_ner1
      f1_ner1 = f1_score(y_true_ner1, y_pred_ner1, average='micro')

      y_true_ner2 = torch.tensor([label for sublist in real_ner2 for label in sublist]).cpu()
      del real_ner2
      y_pred_ner2 = torch.tensor([label for sublist in predicted_ner2 for label in sublist]).cpu()
      del predicted_ner2
      f1_ner2 = f1_score(y_true_ner2, y_pred_ner2, average='micro')


    return avg_loss_ner1, avg_loss_ner2,f1_ner1,f1_ner2
```

```python
import torch.nn.functional as F
log_soft = F.log_softmax
```

```python
class EntityModel(nn.Module):
    def __init__(self, num_ner1, num_ner2):
        super(EntityModel, self).__init__()
        self.num_ner1 = num_ner1
        self.num_ner2 = num_ner2
        self.phobert = transformers.AutoModel.from_pretrained(config.BASE_MODEL_PATH,return_dict=False)
        # for param in self.phobert.parameters():
        #     param.requires_grad = False
        self.bert_drop_1 = nn.Dropout(0.3)
        self.bert_drop_2 = nn.Dropout(0.3)
        self.out_ner1 = nn.Linear(768, self.num_ner1)
        self.out_ner2 = nn.Linear(768, self.num_ner2)
        self.crf_ner1 = CRF(num_ner1, batch_first = True)
        self.crf_ner2 = CRF(num_ner2, batch_first = True)
    def forward(
        self,
        ids,
        mask,
        token_type_ids,
        target_ner1,
        target_ner2
    ):
        o1, _ = self.phobert(
            ids,
            attention_mask=mask,
            token_type_ids=token_type_ids
        )

        bo_ner1 = self.bert_drop_1(o1)
        bo_ner2 = self.bert_drop_2(o1)

        ner1 = self.out_ner1(bo_ner1)
        ner2 = self.out_ner2(bo_ner2)



        if target_ner1 is not None:
            loss_ner1 = -self.crf_ner1(log_soft(ner1, 2), target_ner1, mask=mask.type(torch.uint8), reduction='mean')
            prediction_ner1 = self.crf_ner1.decode(ner1, mask=mask.type(torch.uint8))


        else:
            prediction_ner1 = self.crf_ner1.decode(ner1, mask=mask.type(torch.uint8))
            # return prediction

        if target_ner2 is not None:
            loss_ner2 = -self.crf_ner2(log_soft(ner2, 2), target_ner2, mask=mask.type(torch.uint8), reduction='mean')
            prediction_ner2 = self.crf_ner2.decode(ner2, mask=mask.type(torch.uint8))


        else:
            prediction_ner2 = self.crf_ner2.decode(ner2, mask=mask.type(torch.uint8))


        return prediction_ner1, prediction_ner2,loss_ner1,loss_ner2
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
param_optimizer = list(model.named_parameters())
# no_decay = ["bias", "LayerNorm.bias", "LayerNorm.weight"]
# no_decay = ['bias', 'gamma', 'beta','LayerNorm.weight','LayerNorm.bias']
no_decay = ['bias', 'gamma', 'beta',"LayerNorm.bias", "LayerNorm.weight"]
optimizer_parameters = [
    {
        "params": [
            p for n, p in param_optimizer if not any(
                nd in n for nd in no_decay
            )
        ],
        "weight_decay": 0.001,
    }

]

num_train_steps = int(
    len(sentences_train) / config.TRAIN_BATCH_SIZE * 10
)
optimizer = AdamW(optimizer_parameters, lr=3e-5)
optimizer1 = AdamW(optimizer_parameters, lr=3e-5)
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=0,
    num_training_steps=num_train_steps
)

best_loss = np.inf
# best_acc = 0.0
for epoch in range(10):
  # avg_loss_ner1, avg_loss_ner2,f1_ner1,f1_ner2
    train_ner1_losses, train_ner2_losses,f1_ner1_train,f1_ner2_train = train_fn(
        train_data_loader,
        model,
        optimizer,
        optimizer1,
        device,
        scheduler
    )

    test_ner1_losses, test_ner2_losses,f1_ner1_test,f1_ner2_test = eval_fn(
        valid_data_loader,
        model,
        device
    )

    print(f"Train NER1 Loss = {train_ner1_losses:.4f}")
    print(f"Train NER2 Loss = {train_ner2_losses:.4f}")
    print(f"Valid NER1 Loss = {test_ner1_losses:.4f}")
    print(f"Valid NER2 Loss = {test_ner2_losses:.4f}")

    print(f"Train NER1 F1 = {f1_ner1_train:.4f}")
    print(f"Train NER2 f1 = {f1_ner2_train:.4f}")
    print(f"Valid NER1 f1 = {f1_ner1_test:.4f}")
    print(f"Valid NER2 f1 = {f1_ner2_test:.4f}")


    test=test_ner1_losses + test_ner2_losses
    if test < best_loss:
        torch.save(model.state_dict(), config.MODEL_PATH)
        print("Save model.")
        best_loss = test
# torch.save(model.state_dict(), '/content/drive/MyDrive/Colab Notebooks/test_model/best_model.pth')
```
